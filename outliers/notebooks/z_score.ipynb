## Library imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import describe
from sklearn.decomposition import PCA
```

## One-dimensional Observations

### Normal distribution

```python
p_dist = [3, 0.75]
np.random.seed(100)
target_set = np.random.normal(loc=p_dist[0], scale=p_dist[1], size=100)
outlier_set = np.array([5, 1])
target_set = np.concatenate((target_set, outlier_set), axis=0)
```

```python
# target variable distribution
plt.hist(target_set)
plt.show()

# Z-score works to characterize outlierness
z_target = (target_set - np.mean(target_set))/np.std(target_set)
plt.hist(z_target)
plt.show()
```

### Multinomial distribution

```python
p_dist = [0.05, 0.15, 0.4, 0.3, 0.1]
np.random.seed(100)
sample_set = np.random.multinomial(100, p_dist)
target_set = np.random.permutation([x for i in range(5) for x in [i]*sample_set[i]])
```

```python
np.std(target_set)
```

```python
# response distribution
plt.hist(target_set)
plt.show()

# Z-score still works when unimodal distribution. Perhaps need to use student's t-distribution or discrete distribution
# characterization here to get better bounds.
z_target = (target_set - np.mean(target_set))/np.std(target_set)
plt.hist(z_target)
plt.show()
```

## Two-dimensional observations

```python
mean_1 =  [0, 0]
mean_2 = [10, 10]
cov_1 = np.array([[3, 0], [0, 3]])
cov_2 = np.array([[1, 0], [0, 1]])
np.random.seed(100)
target_set_1 = np.random.multivariate_normal(mean_1, cov_1, 100)
target_set_2 = np.random.multivariate_normal(mean_2, cov_2, 100)
target_set_3 = np.array([[5, 5]])
target_set = np.concatenate((target_set_1, target_set_2, target_set_3), axis=0)
```

```python
# response distribution
plt.scatter(*target_set.T)
```

```python
pca_model = PCA(n_components=1)
target_1dim_set = pca_model.fit_transform(target_set)
```

```python
# response distribution
plt.hist(target_1dim_set)
plt.show()

# Z-score does not work here. Need Gaussian mixture model.
z_target = (target_1dim_set - np.mean(target_1dim_set))/np.std(target_1dim_set)
plt.hist(z_target)
plt.show()
```
